```python
import random
def add_zeros(n, cifras):
    n = str(n)
    while len(n) < cifras:
        n = "0" + n
    return n

class Date():
    def __init__(self, day=1, month=1, year=1):
        self.day = day
        self.month = month
        self.year = year
        if not self.validDate():
            print("¡¡La fecha no es válida!!")

    def __str__(self):
        strFecha = add_zeros(self.day, 2) + "/" + add_zeros(self.month, 2) + "/" + add_zeros(self.year, 4)
        return strFecha
    
    def isLeap(self):
        isLeap=False
        if self.year%4==0:
            if self.year%100==0:
                if self.year%400==0:
                    return True
                else:
                    return False
            else:
                return True
        else:
            return False

    def totalMonthDays(self):
        if self.month in [1, 3, 5, 7, 8, 10, 12]:
            return 31
        elif self.month in [4, 6, 9, 11]:
            return 30
        elif self.month == 2 and self.isLeap():
            return 29
        else:
            return 28

    def validDate(self):
        if self.year < 0:
            return False
        if self.month <= 0 or self.month > 12:
            return False
        if self.day < 0 or self.day > self.totalMonthDays():
            return False
        return True
    
    def monthName(self):
        diasMes=["January", "February", "March", "April", "May", "June", "July", "August", "September", "October", "November", "December"]
        return diasMes[self.month-1]

    @staticmethod
    def areEqual(data1,data2):
        return data1.__str__()==data2.__str__()

    @staticmethod
    def isLater(data1,data2):
        str1 = int(add_zeros(data1.year, 4) + add_zeros(data1.month, 2) + add_zeros(data1.day, 2))
        str2 = int(add_zeros(data2.year, 4) + add_zeros(data2.month, 2) + add_zeros(data2.day, 2))
        return str1 > str2

    @staticmethod
    def isPrevious(data1,data2):
        str1 = int(add_zeros(data1.year, 4) + add_zeros(data1.month, 2) + add_zeros(data1.day, 2))
        str2 = int(add_zeros(data2.year, 4) + add_zeros(data2.month, 2) + add_zeros(data2.day, 2))
        return  str1 < str2

    @classmethod
    def firstDayOfTheYear(cls, urtea):
        if urtea >0:
            return cls(1,1,urtea)
        print("El año introducido no es válido")

    @classmethod
    def lastDayOfTheYear(cls, urtea):
        if urtea >0:
            return cls(31,12,urtea)
        print("El año introducido no es válido")

    def plusDay(self):
        if self.month==12 and self.day==31:
            self.day=1
            self.month=1
            self.year+=1
        elif self.day==self.totalMonthDays():
            self.day=1
            self.month+=1
        else:
            self.day+=1

    def minusDay(self):
        if self.month==1 and self.day==1:
            self.day=31
            self.month=12
            self.year-=1
        elif self.day==1:
            self.month-=1
            self.day=self.totalMonthDays()
        else:
            self.day-=1

    @classmethod
    def copy(cls, fecha):
        copia=cls(fecha.day,fecha.month,fecha.year)
        return copia

    @staticmethod
    def difference(data1,data2):
        if Date.areEqual(data1,data2):
            return 0
        elif Date.isLater(data1,data2):
            count=0
            copia2=Date.copy(data2)
            while not Date.areEqual(data1,copia2):
                count+=1
                copia2.plusDay()
            return count
        else:
            count=0
            copia1=Date.copy(data1)
            while not Date.areEqual(copia1,data2):
                count+=1
                copia1.plusDay()
            return count

    @classmethod
    def randomDate(cls):
        year=random.randint(0,2021)
        month=random.randint(1,12)
        random_date=cls(1,month,year)
        day=random.randint(1,random_date.totalMonthDays())
        random_date.day=day
        return random_date

    @classmethod
    def toDate(cls, strFecha):
        partes = strFecha.split(" ")
        day = int(partes[0])
        year = int(partes[2])
        diasMes = ["January", "February", "March", "April", "May", "June", "July", "August", "September", "October", "November", "December"]
        month = diasMes.index(partes[1]) + 1
        return cls(day, month, year)

```
