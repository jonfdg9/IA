## EJERCICIO 1

Dibuja 100 puntos aleatorios (entre 0 y 100 para ambas coordenadas) con colores y tamaños aleatorios y una transparencia de 0.5.

Tendrás que crear 4 arrays aleatorios de 100 posiciones: uno para las coordenadas x, otro para y, otro para el color de cada punto y un último para el tamaño de cada punto. Para que los puntos no se vean muy pequeños, multiplica el array de tamaños por 10.

Configura el tamaño de la figura a 10 x 10. Para ello, investiga el método .figure() de plt.

Por último, selecciona los colores combinando el array aleatorio y el mapa de color llamado hsv y recuerda mostrar la colorbar.

````python
import numpy as np
import matplotlib.pyplot as plt

# Crear 100 puntos aleatorios para las coordenadas x e y
x = np.random.uniform(0, 100, 100)
y = np.random.uniform(0, 100, 100)

# Crear colores aleatorios en el rango [0, 1] para el mapa de color hsv
color_values = np.random.rand(100)

# Crear tamaños aleatorios y multiplicarlos por 10
sizes = np.random.randint(1, 11, 100) * 10

# Configurar el tamaño de la figura a 10x10
plt.figure(figsize=(10, 10))

# Crear el gráfico de dispersión con el colormap hsv
plt.scatter(x, y, c=color_values, s=sizes, alpha=0.5, cmap='hsv')

# Añadir barra de color
plt.colorbar()

# Configurar el título y las etiquetas
plt.title('Puntos aleatorios con colores y tamaños aleatorios')
plt.xlabel('Coordenada X')
plt.ylabel('Coordenada Y')

# Establecer los límites de los ejes
plt.xlim(0, 100)
plt.ylim(0, 100)

# Mostrar el gráfico
plt.grid(True)
plt.show()
````

## EJERCICIO 2

Crea una función que dado un color en forma RGB lo transforme en su forma hexadecimal.

````python
def rgb_to_hex(rgb):
    """
    Convierte un color en formato RGB a su forma hexadecimal.

    Parámetros:
    rgb (tuple): Un tuple que contiene los valores RGB (r, g, b), donde r, g y b están en el rango [0, 255].

    Retorna:
    str: El color en formato hexadecimal.
    """
    # Asegurarse de que los valores están en el rango correcto
    if not all(0 <= x <= 255 for x in rgb):
        raise ValueError("Los valores RGB deben estar en el rango [0, 255]")

    # Convertir a hexadecimal y formatear
    return '#{:02x}{:02x}{:02x}'.format(*rgb)

# Ejemplo de uso
color_rgb = (255, 105, 180)  # RGB para un color rosa
color_hex = rgb_to_hex(color_rgb)
print(color_hex)  # Salida: #ff69b4
````

## EJERCICIO 3

Dibuja la función sin(x)/x de color #f06741, la función (2sin(x))/x de color #a4f041 y la función -sin(x)/x de color #8d41f0 donde x ∈ ```[-5π, 5π]``` en el mismo plot.

Crea el array x de tamaño 100 con el método .linspace() de numpy.

Haz que el tamaño de la figure sea 8 x 6.

````python
import numpy as np
import matplotlib.pyplot as plt

x=np.linspace(-5*np.pi,5*np.pi,num=100)

f1=np.sin(x)/x
f2=2*np.sin(x)/x
f3=-np.sin(x)/x


plt.figure(figsize=(8, 6))
plt.plot(x , f1 , color="#f06741", label='sin(x)/x')
plt.plot(x , f2 , color="#a4f041", label='(2sin(x))/x')
plt.plot(x , f3 , color="#8d41f0", label='-sin(x)/x')

plt.title('Gráfico de las funciones')
plt.xlabel('x')
plt.ylabel('f(x)')

plt.axhline(0, color='black', lw=0.5, ls='--')
plt.axvline(0, color='black', lw=0.5, ls='--')

plt.legend()
plt.show()

````
