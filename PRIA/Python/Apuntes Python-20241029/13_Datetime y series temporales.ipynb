## Tiempo con `datetime`

El módulo datetime de `Python` proporciona clases para manipular fechas y horas. Este módulo es muy útil cuando se trata de trabajar con operaciones relacionadas con el tiempo, como la diferencia entre fechas, el formateo de fechas, etc.

Una fecha en `Python` no es un tipo de dato por si mismo, pero podemos trabajar con estos objetos tan peculiares con el módulo `datetime`

```python
import datetime as dt
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

Para obtener la fecha y hora actual, podemos usar el método .datetime.now()

```python
today = dt.datetime.now()
print(today)
```

```python
type(today)
```

Observación. Dentro del módulo datetime, existe la clase datetime. Para evitar confusiones hemos renombrado al módulo datetime por dt

En el resultado anterior vemos que se nos muestra el año, mes, día, hora, minuto, segundo e incluso microsegundo.

Podemos acceder a toda esa información con los atributos

* `.year`: año
* `.month`: mes
* `.day`: día
* `.hour`: hora
* `.minute`: minuto
* `.second`: segundo

```python
print("Año:", today.year)
print("Mes:", today.month)
print("Día:", today.day)
print("Hora:", today.hour)
print("Minuto:", today.minute)
print("Segundo:", today.second)
print("Microsegundo:", today.microsecond)
```

### Creando objetos `datetime`

Para crear objetos `datetime` podemos usar el método `.datetime()`. 

Este método toma como parámetros el año (`year`), mes (`month`), día (`day`), hora (`hour`), minuto (`minute`), segundo (`second`) y microsegundo (`microsecond`). Los 3 primeros, relacionados con la fecha, son parámetros obligatorios. Los parámetros restantes, los relacionados con el tiempo, son opcionales.

```python
d = dt.datetime(2020, 9, 22, 12, 30)
print(d)
```

Observación. El método .datetime() toma un parámetro adicional, tzinfo, que también es opcional. A este parámetro se le suministra la zona horaria. Por defecto toma el valor None.

Es hora de hablar del método opuesto a .strftime(). Se trata del método .strptime(), que nos permite crear un objeto datetime a partir de un string. No obstante, como el string puede tener cualquier formato, habrá que indicarle por parámetro a .strptime() cuál estamos usando.

```python
dt.datetime.strptime("September 22, 2020", "%B %d, %Y")
```

```python
dt.datetime.strptime("22/09/2020", "%d/%m/%Y")
```

```python
dt.datetime.strptime("22 Sep 20", "%d %b %y")
```

```python
dt.datetime.strptime("22-09-2020 12:30:00", "%d-%m-%Y %H:%M:%S")
```

```python
dt.datetime.strptime("22 Sep 2020, 12:30PM", "%d %b %Y, %I:%M%p")
```

¡Cuidado! El patrón debe coincidir a la perfección. Eso incluye las comas, los : y los espacios en blanco.

### Uso de timedelta para operaciones de tiempo:

Timedelta es una clase del módulo datetime en Python que se utiliza para representar diferencias de tiempo, como el intervalo entre dos fechas o tiempos. Puede ser muy útil para sumar o restar tiempo a objetos datetime, o para medir duraciones.

```python
from datetime import datetime, timedelta

now = datetime.now()
print("Fecha y hora actuales:", now)

# Sumar 10 días
future_date = now + timedelta(days=18)
print("10 días en el futuro:", future_date)

# Restar 2 horas
past_time = now - timedelta(hours=2)
print("2 horas en el pasado:", past_time)
```

## Series temporales

Las series temporales son secuencias de datos recolectados, registrados o medidos en puntos sucesivos en el tiempo, a menudo con intervalos uniformes. Son ampliamente utilizadas en aplicaciones que involucran análisis financiero, economía, climatología, ciencia de datos y más. El análisis de series temporales implica entender patrones como tendencias, estacionalidad, autocorrelación, y realizar pronósticos futuros basados en datos históricos.

Pandas es la biblioteca fundamental para manipular datos en formato tabular, incluidos los datos de series temporales. Ofrece una gran cantidad de funciones para manejar fechas y tiempos, resampling (reescalar intervalos de tiempo), interpolación, alineación de datos por fechas, etc.

##### `Tendencia`: Es el patrón a largo plazo en los datos de la serie temporal. Si los datos están creciendo o disminuyendo con el tiempo, se dice que la serie tiene una tendencia.

```python
# Crear una serie temporal con tendencia
np.random.seed(42)
n = 100
tendencia = np.linspace(0, 10, n)  # Tendencia creciente
ruido = np.random.normal(0, 1, n)  # Ruido aleatorio
datos = tendencia + ruido

# Convertir los datos a una serie de pandas
fechas = pd.date_range('2023-01-01', periods=n)
serie_temporal = pd.Series(datos, index=fechas)

# Graficar la serie temporal
serie_temporal.plot(title='Serie Temporal con Tendencia')
plt.show()
```

##### `Estacionalidad`: Son patrones que se repiten en intervalos regulares en los datos. Por ejemplo, las ventas de productos en un año podrían aumentar en los meses de verano y reducirse en invierno.

```python
# Crear una serie temporal con estacionalidad
n = 100
periodo = 12  # Cada 12 puntos se repite el ciclo
estacionalidad = 10 * np.sin(np.linspace(0, 3 * np.pi, n))  # Patrón estacional
ruido = np.random.normal(0, 1, n)
datos_estacionales = estacionalidad + ruido

# Convertir los datos a una serie de pandas
serie_estacional = pd.Series(datos_estacionales, index=fechas)

# Graficar la serie temporal con estacionalidad
serie_estacional.plot(title='Serie Temporal con Estacionalidad')
plt.show()
```

##### `Ruido`: Son fluctuaciones aleatorias en los datos que no siguen ningún patrón sistemático.

```python
# Crear una serie temporal con solo ruido
ruido_puro = np.random.normal(0, 1, n)

# Convertir los datos a una serie de pandas
serie_ruido = pd.Series(ruido_puro, index=fechas)

# Graficar la serie temporal con solo ruido
serie_ruido.plot(title='Serie Temporal con Ruido')
plt.show()
```

#####  `Autocorrelación`: Es la correlación de una serie temporal consigo misma en diferentes puntos en el tiempo. Es útil para entender cómo los valores pasados afectan los futuros.

```python
from pandas.plotting import autocorrelation_plot

# Crear una serie temporal con autocorrelación
datos_autocorrelacion = tendencia + 0.5 * np.roll(tendencia, 1) + ruido  # Introducimos autocorrelación

# Convertir los datos a una serie de pandas
serie_autocorrelacion = pd.Series(datos_autocorrelacion, index=fechas)

# Graficar la autocorrelación
autocorrelation_plot(serie_autocorrelacion)
plt.title('Autocorrelación de la Serie Temporal')
plt.show()
```

##### `Diferenciación`: Proceso de restar los valores actuales de los valores anteriores para estabilizar la media en una serie temporal no estacionaria.

```python
# Crear una serie temporal con tendencia
n = 100
tendencia_lineal = np.linspace(0, 10, n)
ruido_diferencia = np.random.normal(0, 1, n)
datos_con_tendencia = tendencia_lineal + ruido_diferencia

# Convertir los datos a una serie de pandas
serie_con_tendencia = pd.Series(datos_con_tendencia, index=fechas)

# Aplicar diferenciación
serie_diferenciada = serie_con_tendencia.diff().dropna()

# Graficar antes y después de la diferenciación
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
serie_con_tendencia.plot(title='Serie con Tendencia')
plt.subplot(1, 2, 2)
serie_diferenciada.plot(title='Serie Diferenciada')
plt.show()
```

Este método genera un rango de fechas (índice temporal) que puede ser usado para crear series temporales o para establecer un índice en un DataFrame. Es extremadamente útil para trabajar con datos que tienen intervalos regulares (diarios, semanales, mensuales, etc.).

 El metodo `pd.to_datetime()`: Convierte cadenas de texto a formato de fecha. Este método convierte diferentes tipos de entradas (cadenas de texto, listas o enteros) en objetos datetime que pueden ser usados como índices o columnas de fechas en los DataFrames. Esto es esencial cuando tienes datos con fechas en formato de texto y necesitas tratarlos como objetos temporales.

Parámetros:
* arg: El dato a convertir (puede ser una cadena, lista, Series, etc.).
* format: El formato de la fecha (por ejemplo, '%Y-%m-%d'). Útil para acelerar la conversión si conoces el formato.
* errors: Controla qué sucede si hay errores de conversión ('raise', 'coerce', 'ignore').
* infer_datetime_format: Si es True, intentará inferir el formato automáticamente.

```python
# Convertir una cadena de texto a formato datetime
fecha = pd.to_datetime('2024-01-01')
print(fecha)

# Convertir una lista de fechas en cadenas a datetime
fechas = pd.to_datetime(['2024-01-01', '2024-01-05', '2024-01-10'])
print(fechas)
```

 El metodo `pd.resample()`: Cambiar la frecuencia temporal de los datos en una serie o un DataFrame. Es similar al método groupby(), pero en lugar de agrupar por categorías, se agrupa por períodos de tiempo (frecuencias temporales). Por ejemplo, convertir datos diarios en mensuales, o desagregar datos anuales a trimestrales.

```python
# Crear una serie temporal diaria
rng = pd.date_range('2024-01-01', periods=30, freq='D')
data = pd.Series(np.random.randn(len(rng)), index=rng)
print(data)

# Resamplear los datos a frecuencia mensual (M) sumando los valores
data_mensual = data.resample('M').sum()
print(data_mensual)
```

### Suma de series

Las series se pueden sumar con el elemento +, esto implica sumar las dos series elemento a elemento, es decir, sumará los valores de las series con los mismos índices. Si ambas series tienen los mismos índices, se sumarán los valores correspondientes en cada índice. Si los índices no coinciden completamente, pandas tratará de alinear las series en base a los índices. Para los índices que están en ambas series, sumará los valores correspondientes. Para los índices que no existen en ambas series, se asignará NaN (Not a Number), que indica valores faltantes.
