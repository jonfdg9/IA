## Ejercicio 3 - Creación y manejo de una clase Alumno - 6 Puntos

Realizar un programa que conste de una clase llamada Alumno que tenga como atributos el nombre y la nota del alumno. Definir los siguientes métodos:

1. Definir la clase Alumno:

* **Atributos:**
    * nombre: nombre del alumno.
    * nota: calificación obtenida por el alumno.
* **Métodos:**
    * Método constructor: Inicializa el nombre y la nota del alumno.
    * Sobrescritura del método to_string: Devuelve una representación amigable de los atributos del alumno.
    * Método de instancia `mostrar_detalle()`: Imprime los atributos con la frase: "El alumno XXX ha sacado la nota YYY."
    * Método de clase `asignar_nota_aleatoria(cls, nombre)`: Crea un nuevo alumno con una nota aleatoria entre 0 y 10.
    * Método estático `comparar_notas(alumno1, alumno2)`: Compara las notas de dos alumnos y devuelve un mensaje indicando cuál tiene la mayor nota, o si están empatados.
2. Crear dos objetos:
    * Uno utilizando el constructor normal.
    * Otro utilizando el método de clase.
3. Utilizar los métodos:
    * Mostrar los detalles de ambos alumnos.
    * Comparar sus notas con el método estático.


```python
class Alumno:
    def __init__(self, nombre, nota):
        self.nombre=nombre
        self.nota=nota

    def __str__(self):
        return f"Alumno: {self.nombre}, Nota: {self.nota}"
    
    def mostrar_detalle(self):
        print(f"El alumno {self.nombre} ha sacado la nota {self.nota}.")

    @classmethod
    def asignar_nota_aleatoria(cls,nombre):
        nota_aleatoria= np.random.randint(0,11)
        return cls(nombre, nota_aleatoria)
    
    @staticmethod
    def comparar_notas(alumno1, alumno2):
        if alumno1.nota > alumno2.nota:
            return f"{alumno1.nombre} tiene una nota mayor {alumno1.nota} que {alumno2.nombre} {alumno2.nota}."
        elif alumno1.nota < alumno2.nota:
            return f"{alumno2.nombre} tiene una nota mayor {alumno2.nota} que {alumno1.nombre} {alumno1.nota}."
        else:
            return f"{alumno1.nombre} y {alumno2.nombre} tienen la misma nota {alumno1.nota}."
        
alumno1=Alumno("Jon",7)
alumno2=Alumno.asignar_nota_aleatoria("Joritz")

print(alumno1)
alumno2.mostrar_detalle()   

Alumno.comparar_notas(alumno1, alumno2)

```

```python
df = pd.read_csv("Recursos/06-vg_sales.csv")

# Creamos el objeto figure
fig = plt.figure(constrained_layout = True)

# Configuramos que tendrá 3 filas y 4 columnas
gs = fig.add_gridspec(3, 3)

# Creamos los subplots. En este caso 3
# * Ocupa todas las filas y todas las columnas salvo la última
fig_ax1 = fig.add_subplot(gs[:-1, :])
#fig_ax1.set_title("Tarta")

wii_df= df[df['Name'] == 'Wii Sports']

colors = ["#fbb4ae", "#b3cde3", "#ccebc5", "#decbe4"]
labels= ['NA_Sales', 'EU_Sales', 'JP_Sales', 'Other Sales']

data = []
data.append(wii_df.iat[0,6])
data.append(wii_df.iat[0,7])
data.append(wii_df.iat[0,8])
data.append(wii_df.iat[0,9])

fig_ax1.pie(
    data,
    labels=labels,
    colors=colors,
    autopct="%1.1f%%",   # Mostrar porcentajes con formato "%1.1f%%"
    startangle=90,       # Empieza en el ángulo 90
    shadow=True,         # Incluir sombreado
    explode=[0,0,0.5,0],    # Separación de 0.5 al sector más pequeño
)

fig_ax1.set_title("Wii Sports")


# * Ocupa  la ultima fila y dos columna
fig_ax2 = fig.add_subplot(gs[2:3, :-1])
fig_ax2.set_title("EU vs NA sales")
sns.scatterplot(data = df, x = "NA_Sales", y = "EU_Sales",)
fig_ax2.set_title("Propina vs. Total pagado")
fig_ax2.set_xlabel("NA_Sales")
fig_ax2.set_ylabel("EU_Sales")


# * Ocupa la última fila y la última columna
fig_ax3 = fig.add_subplot(gs[-1, -1])

consolas = df['Platform'].unique()
countConsola=df['Platform'].unique()

fig_ax3.bar(consolas, countConsola, width = 0.5)
fig_ax3.set_xlabel("Consolas")
fig_ax3.set_ylabel("Ventas")
fig_ax3.set_title("Ventas por consola")

plt.show()
```
